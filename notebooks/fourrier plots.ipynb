```python
img = cv2.imread("../moonlanding.png", cv2.IMREAD_GRAYSCALE)
img = pad_image(img)
```

```python
result = FFT2(img)
```

```python
temp = result.real
```

```python
plt.figure(figsize=(20, 20))

plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title("Original image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(temp, norm=colors.LogNorm(vmin=5))
plt.title("Transformed image"), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(121), plt.imshow(np.fft.fft2(img).real, norm=colors.LogNorm(vmin=5))
plt.title("Numpy built-in fourier transform"), plt.xticks([]), plt.yticks([])
```
