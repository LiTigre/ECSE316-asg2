```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, '../src')

from my_fourrier import *
from helpers import *
```

```python
X = [1<<exponent for exponent in range(13)]

for i, array_size in enumerate([2048, 1024, 512, 256, 128, 64]):
    run_time = []
    test_array = np.random.random(array_size)
    
    for min_size in X[:-(i + 1)]:
        
        start = time.time() * 1000
        FFT(test_array, size = min_size)
        end = time.time() * 1000

        run_time.append(end - start)
        
    plt.plot(X[:-(i + 1)], run_time, label = "size of input - " + str(array_size))

plt.legend(loc = 2)
plt.title("Run-time of FFt with different thresholds")
plt.xlabel("threshold size")
plt.ylabel("runtime (ms)")

plt.show()
```
